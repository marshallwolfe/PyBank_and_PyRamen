```python
# Import libraries
import csv
from pathlib import Path

# Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('Resources/menu_data.csv')
sales_filepath = Path('Resources/sales_data.csv')

# Initialize list objects to hold our menu and sales data
menu = []
sales = []
```

```python
# Read in the menu data into the menu list
with open(menu_filepath, 'r') as menu_data:
    menu_reader = csv.reader(menu_data, delimiter=',')

    # Skip the header row
    # menu_header = next(menu_reader)

    for row in menu_reader:
        # print(row)
        menu.append(row)

print(menu[0])

# Read in the sales data into the sales list
with open(sales_filepath, 'r') as sales_data:
    sales_reader = csv.reader(sales_data, delimiter=',')
    
    # Skip the header row
    # sales_header = next(sales_reader)

    for row in sales_reader:
        # print(row)
        sales.append(row)

print(sales[0])
```
